## 코드

```python
# 백트래킹은 모든 가능한 경우의 수 중에서 특정한 조건을 만족하는 경우만 살펴보는 것
# 모든 경우의 수 = 모든 연산기호의 조합
# 탈출 조건 = 연산기호를 모두 다 사용하면 (depth가 n이되면)
```

```python
n = int(input())
num = list(map(int, input().split()))
oper = list(map(int, input().split()))

big = -1e9
small = 1e9
```

```python
def dfs(plus, minus, mul, div, depth, result):
    global big, small
    if depth == n: # 탈출조건 = 연산기호를 모두 사용하면
        big = max(result, big)
        # 가장 큰 결과
        small = min(result, small)
        # 가장 작은 결과
        return


    if plus:
        dfs(plus - 1, minus, mul, div, depth + 1, result + num[depth])
    if minus:
        dfs(plus, minus - 1,  mul, div, depth + 1, result - num[depth])
    if mul:
        dfs(plus, minus, mul - 1, div, depth + 1, result * num[depth])
    if div:
        dfs(plus, minus, mul, div- 1, depth + 1, int(result / num[depth]))

dfs(oper[0], oper[1], oper[2], oper[3], 1, num[0])
print(big)
print(small)
```
