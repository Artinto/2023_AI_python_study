```python
n = int(input())
num_list = list(map(int, input().split()))
arithmetic_list = list(map(int, input().split()))
max_result = -1e9
min_result = 1e9

def backtracking(index, value, add, sub, mul, div):
  global max_result, min_result
  if index == n:
    max_result = max(max_result, value)
    min_result = min(min_result, value)
    return
  if add > 0:
    backtracking(index + 1, value + num_list[index], add - 1, sub, mul, div)
  if sub > 0:
    backtracking(index + 1, value - num_list[index], add, sub - 1, mul, div)
  if mul > 0:
    backtracking(index + 1, value * num_list[index], add, sub, mul - 1, div)
  if div > 0:
    backtracking(index + 1, int(value / num_list[index]), add, sub, mul, div - 1)

backtracking(1, num_list[0], *arithmetic_list) # 처음 값은 value로 들어갓으므로 index는 1.
print(max_result)
print(min_result)
```
