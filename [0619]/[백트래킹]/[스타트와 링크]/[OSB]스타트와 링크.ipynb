```python
def backtracking(index, team_start, team_link, n, total):
  global min_result
  if index == n: # 다 확인해봤을 때 종료
    if len(team_start) != n // 2 or len(team_link) != n // 2:
      return
    start_sum = 0 # start팀 합
    link_sum = 0 # link팀 합
    for i in range(n//2):
      for j in range(n//2):
        start_sum += total[team_start[i]][team_start[j]]
        link_sum += total[team_link[i]][team_link[j]]
    value = abs(start_sum - link_sum) # 능력치 차이 비교
    min_result = min(min_result, value)
    return
  backtracking(index + 1, team_start + [index], team_link, n, total) # start team에  대해 백트래킹
  backtracking(index + 1, team_start, team_link + [index], n, total) # link team에 대해 백트래킹

n = int(input())
total = [list(map(int, input().split())) for _ in range(n)]
min_result = 1e9
backtracking(0, [], [], n, total)
print(min_result)
```
