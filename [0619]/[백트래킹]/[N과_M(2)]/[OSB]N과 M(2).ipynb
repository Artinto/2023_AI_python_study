* 재귀함수 이용

```python
N, M = map(int, input().split())
num_list = [i+1 for i in range(N)]

def combination(num_list, m):
  answer = []  # 정답 출력을 위한 리스트 선언.
  if m > len(num_list):  # 더 많이 뽑는 경우 끝내버림.
    return answer
  if m == 1:  # 만약 뽑는 숫자가 1인 경우
    for i in num_list:
      answer.append([i])  # 리스트의 각 원소를 하나의 원소로 갖는 리스트로 변환
  elif m > 1:  # 뽑는 숫자 2 이상인 경우
    for i in range(len(num_list)-m+1):
      temp = [num for num in num_list[i+1:]]  # 선택한 요소의 다음 원소부터 리스트 끝까지.중복을 피하기 위해
      for j in combination(temp, m - 1):  # 재귀 호출을 통해 m-1의 길이를 구함.
        answer.append([num_list[i]] + j)  # num_list[i]와 재귀적인 조합 결합
  return answer

result = combination(num_list, M)
for i in result:
  for v in i:
    print(v, end=' ')
  print('')
```

* 모듈 이용.

```python
from itertools import combinations
n,m = map(int, input().split())

result = list(combinations(range(1, n+1), m))
for i in result:
  for v in i:
    print(v, end = ' ')
  print("")
```
