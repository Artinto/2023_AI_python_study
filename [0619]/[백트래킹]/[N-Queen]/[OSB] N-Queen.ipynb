# 풀이
* 퀸은 같은 줄과 같은 대각선 상에 있으면 안된다.
* 2차원 배열보다 1차원 배열로 해서 대각선 여부를 판단


```python
import sys
N = int(sys.stdin.readline().rstrip())

def backtracking(x):
  global result
  if x == N:
    result += 1
  else:
    for i in range(N): # 0부터 N 전까지 옮겨가면서 퀸 놓기
      row[x] = i
      if adjacent(x): # 행,열,대각선 체크함수 true이면 백트래킹 안하고 계속 진행
        backtracking(x + 1)

def adjacent(x): # 같은 선상일 때 판단
  for i in range(x):
    if row[x] == row[i] or abs(row[x] - row[i]) == x - i: # 열이 같거나 같은 대각선 상이면 False
      return False
  return True

row = [0] * N
result = 0
backtracking(0)
print(result)
```
