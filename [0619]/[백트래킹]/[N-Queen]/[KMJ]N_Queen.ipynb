## 코드

```python
# 퀸을 놓을 수 없는 칸을 체크할 때, 8개의 모든 방향에 대해서 체크할 필요 없이 아래쪽, 왼쪽 아래 대각선, 오른쪽 아래 대각선의 3 방향만 체크
# 탈출 조건 = 어느 줄이라도 퀸을 놓을 수 없는 줄이 존재하면 바로 종료, 마지막 줄에 다다르면 종료
```

```python
n = int(input())
result = 0
visited = [[False for _ in range(n)] for _ in range(n)] # N*N 체스판 list 생성

def visit(x, y, in_visited): # 퀸의 위치 체크
    tmp_visited = [visi[:] for visi in in_visited]
    for i in range(1, n-x):
        tmp_visited[x+i][y] = True  # 아래 방향 체크
        if 0 <= y-i < n:
            tmp_visited[x+i][y-i] = True    # 왼쪽 아래 대각선 체크
        if 0 <= y+i < n:
            tmp_visited[x+i][y+i] = True    # 오른쪽 아래 대각선 체크
    return tmp_visited


def recursion(q, _visited):     # q번째 줄에 퀸을 둘 수 있는 경우들을 확인하는 재귀함수
    global result
    for idx in range(q, n):
        if sum(_visited[idx]) == n: # 탈출조건 = 퀸을 놓을 수 없는 줄 존재
            return 0
    if q == (n-1): # 탈출조건 = 마지막 줄에 도착
        result += n - sum(_visited[q])
        return 0

    for i in range(n):
        if not _visited[q][i]:   # 퀸을 둘 수 있는 경우
            tmp = visit(q, i, _visited)     # 퀸을 뒀을 때 불가능한 칸들 체크
            recursion(q+1, tmp)     # 그 다음 줄에 대해 재귀 호출
        # 재귀호출 종료 후 퀸을 둘 수 있는 다른 경우에 대해 체크

recursion(0, visited)   # 0번째 줄부터 탐색 시작
print(result)
```
