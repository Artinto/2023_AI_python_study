## 코드

```python
# 백트래킹은 모든 가능한 경우의 수 중에서 특정한 조건을 만족하는 경우만 살펴보는 것
# 모든 경우의 수 = n개의 수 중 m개로 만들 수 있는 모든 수 출력
# 탈출 조건 =
```

```python
n, m = map(int, input().split())

s = []

def dfs(start):
    if len(s) == m:
        print(" ".join(map(str, s)))
        return

    # for 문으로 반복
    for i in range(start, n + 1):
            s.append(i)
            dfs(i)
            s.pop()
dfs(1)
```
