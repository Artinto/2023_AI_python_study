* 재귀함수 이용

```python
N, M = map(int, input().split())
num_list= [i+1 for i in range(N)]
def permutation(num_list, m):
  answer = []  # 정답 출력을 위한 리스트 선언.
  if m > len(num_list): #  더 많이 뽑는 경우 끝내버림.
      return answer
  if m == 1:  # 만약 뽑는 숫자가 1인 경우
    for i in num_list:
      answer.append([i])  # 리스트의 각 원소를 하나의 원소로 갖는 리스트로 변환
  elif m > 1: # 뽑는 숫자 2 이상이 ㄹ경우.
    for i in range(len(num_list)):
        temp = [i for i in num_list]
        temp.remove(num_list[i])  # num_list에서 i번째 원소를 제거한 나머지 원소들의 리스트 즉 선택한 요소를 제외한 나머지 숫자.
        for j in permutation(temp, m - 1): # 재귀 호출를 통해 m-1의 길이를 구함.
          answer.append([num_list[i]] + j)
    return answer
permutation(num_list, M)
for i in permutation(num_list, M):
  for v in i:
    print(v, end = ' ')
  print('')
```

* 모듈 이용시

```python
from itertools import permutations
n,m = map(int, input().split())

result = list(permutations(range(1, n+1), m))
for i in result:
  for v in i:
    print(v, end = ' ')
  print("")
```
