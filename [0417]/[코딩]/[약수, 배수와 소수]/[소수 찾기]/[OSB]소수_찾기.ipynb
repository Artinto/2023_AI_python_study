## 풀이
* 소수
  * 본인과 1일 제외한 약수가 X
  * 들어온 수의 반까지만 수로 나눠서 나눠지는지 확인
    * 나눠지면 break형식

```python
try_num = int(input())
in_num =list(map(int, input().split( )))
v = 0
for value in in_num:
  temp = value//2
  for i in range(2, temp+1):
    if value % i == 0:
      v += 1
      break
    else:
      continue
# 1일 경우 예외처리
if 1 in in_num:
  v+=1
print(try_num - v)
```
