## 풀이
* 약수 구하기
  * 해당 수를 1부터 해당 수까지 나누기 진행.
  * 0일경우를 순서대로 리스트에 저장

```python
N, K = map(int, input().split(" "))
divisor_list = [] # 약수 리스트
for num in range(1, N+1):
  if N % num == 0:
    divisor_list.append(num)
try:
  print(divisor_list[K-1])
except:
  print(0)
```

```python
N, K = map(int, input().split())
divisor_list = [num for num in range(1, N+1) if N % num == 0]

if len(divisor_list) < K:
    print(0)
else:
    print(divisor_list[K-1])
```
