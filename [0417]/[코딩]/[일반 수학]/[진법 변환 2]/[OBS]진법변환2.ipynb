## 풀이
* N : 10진법 수
* B : 변경할 진법 수
* 진법 변환 과정
  * 11진법부턴 최대값 A
  * 12진법 최대값 B
  * 36진법 최대값 Z
* 아스키코드 이용  
image.png

```python
asci = [chr(i) for i in range(65, 91)]
num_dict = {}
for i in range(10):
  num_dict[i] = str(i)
num = 10
for v in asci:
  num_dict[num] = v
  num+=1
N, B = map(int, input().split())

r_list = [] # 나머지 저장 리스트
# 무한 반복
while(1):
  r = N % B # 나머지
  r_list.append(r)
  if N < B: # 몫이 없을 때
    break
  N = N // B # 목
  
# dictionary에서 value 값 꺼내기
for i, v in enumerate(r_list):
  r_list[i] = num_dict[v]
print("".join(r_list))
```

## 풀이2 
* divmod 사용
  * 몫과 너머지를 반환
* 딕셔너리 선언 없이 사용.
  * ord 사용
    * 해당 아스키코드값을 가져옴

```python
N, B = map(int, input().split())
ans = ''
while N:
  N, r = divmod(N, B) # 몫과 나머지 
  if r < 10: # 10보다 작을때는 그냥 숫자 -> 문자열로 더함
    ans += str(r)
  else:
    ans += chr(ord('A') + r - 10) # 대문자 A의 코드값을 가져와 10부터 A가 할당되도록 함.
ans = ans[::-1] # 순서 뒤집기
print(ans)
```
