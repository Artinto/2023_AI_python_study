## 풀이
* 처음 입력받는 수는 색종이의 수
* 색종이의 위치(해당 점에서 오른쪽으로 10, 위로 10)
* 색종이의 크기는 $10*10$의 정사각형의 모양
  * 겹치는 면적 x좌표가 작은거+10 - 큰거
  * 겹치는 면적 y좌표가 작은거+10 - 큰거
* 색종이 수 * 100 - 겹치는 면적

```python
num = int(input()) # 색종이 수
x_list = []
y_list = []

sum = 100*num
for i in range(num):
  x, y = map(int, input().split(" ")) # 위치 입력 받기
  x_list.append(x)
  y_list.append(y)
  # 겹친 면적 계산
  for I, X in enumerate(x_list):
    Y = y_list[I]
    X_e, Y_e = abs(X-x), abs(Y-y)
    if i == I: # 같은 색종이에 대한 비교X
      continue
    elif X_e >10:
      continue
    elif Y_e >10: # 겹치지 않을경우
      continue
    else:
      sum -= (10 -abs(X-x)) * (10-abs(Y-y))
print(sum)
```

## 풀이2
* 흰 도화지 위에서
* 스탬프로 찍듯이 찍은 영역을 1로 초기화 하자.


```python
num = int(input()) # 색종이 수
board = [[0]*100 for x in range(100)] #list comprehesion을 이용해 도화지 생성
sum = 0

for i in range(num):
    x, y = map(int, input().split(" ")) # 위치 입력 받기
    # 각 색종이가 차지하는 영역을 1로 표시
    for j in range(x, x+10):
        for k in range(y, y+10):
            board[j][k] = 1

# 1로 표시된 부분들을 모두 더함.
for i in range(100):
    for j in range(100):
        if board[i][j] == 1:
            sum += 1
print(sum)
```
