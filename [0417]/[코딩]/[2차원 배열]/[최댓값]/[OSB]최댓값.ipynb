* 최대값이 두 개 이상인 경우 그 중 한 곳의 위치를 출력.
  * 최대값의 위치도 리스트로 저장
  * 결과를 랜덤을 출력

```python
import random
columns_list = []
lows_list = []
max_num = 0
for i in range(9):
  a = list(map(int, input().split(" "))) # 한 줄을 int형으로 받아와 list로 선언
  for j in range(len(a)):
    if max_num < a[j]:
      max_num = a[j]
      columns = j+1
      lows =  i+1
      # max 값 변경으로 인한 위치 리스트들 초기화
      columns_list = []
      lows_list = []
      columns_list.append(columns)
      lows_list.append(lows)

    elif max_num == a[j]:
      # column과 low 값 변경 후 list에 추가
      columns = j+1
      lows =  i+1
      columns_list.append(columns)
      lows_list.append(lows)
# 랜덤수 출력
rn = random.randrange(0,len(columns_list)) # 0이상 columns_list의 수 미만의 난수 추출 
print(max_num)
print(lows_list[rn], columns_list[rn])
```
