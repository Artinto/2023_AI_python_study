```python
# 상근이의 숫자 카드 개수 및 숫자 카드 입력 받기
N = int(input())
N_set = set(map(int, input().split(' ')))
# 비교할 숫자 입력 받기
M = int(input())
M_list = list(map(int, input().split(' ')))

# 비교
Answer = ['1' if i in N_set  else '0' for i in M_list]
print(" ".join(Answer))
```

* 두번째 해결 방법
  * 이분 탐색 알고리즘 사용.
  * 재밌는 점.
    * 사용된 메모리는 set보다 작지만, 시간은 오래걸림.

```python
# 상근이의 숫자 카드 개수 및 숫자 카드 입력 받기
N = int(input())
N_list = list(map(int, input().split(' ')))
N_list.sort()  # 이분 탐색을 위해 정렬

# 비교할 숫자 입력 받기
M = int(input())
M_list = list(map(int, input().split(' ')))

# 이분 탐색 함수 정의
def binary_search(arr, target):
    left, right = 0, len(arr) - 1 # 맨 왼쪽과 맨 오른쪽의 index값 설정
    while left <= right: # 
        mid = (left + right) // 2 # 가운데 지점 정의
        if arr[mid] == target: # 
            return True
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return False

# 비교
Answer = ['1' if binary_search(N_list, i) else '0' for i in M_list]
print(" ".join(Answer))
```
