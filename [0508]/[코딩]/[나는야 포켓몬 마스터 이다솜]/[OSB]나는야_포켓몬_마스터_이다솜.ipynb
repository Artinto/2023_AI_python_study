```python
N, M = map(int, input().split(" "))
# 도감 채우기
pokemon_dict1 = {}
pokemon_dict2 = {}
for i in range(1, N+1):
  name = input()
  pokemon_dict1[name] = i # key값으로 이름, value로 번호를 넣어줌.
  pokemon_dict2[i] = name # key값으로 번호, value로 i를 넣어줌.

answer_list =[]
for i in range(M): # 문제 풀기
  try:
    problem = input() # 숫자로 들어올 때만 실행.
    problem = int(problem)
    answer_list.append(pokemon_dict2[problem])
  except:
    answer_list.append(str(pokemon_dict1[problem])) # 문자로 들어올때만 실행
print('\n'.join(answer_list))
```

* try ... except문으로 인한 시간 초과 오류
* 입력 값이 숫자인 경우 일떄
  * isdigit() 사용

```python
N, M = map(int, input().split(" "))
# 도감 채우기
pokemon_dict1 = {}
pokemon_dict2 = {}
for i in range(1, N+1):
  name = input()
  pokemon_dict1[name] = i # key값으로 이름, value로 번호를 넣어줌.
  pokemon_dict2[i] = name # key값으로 번호, value로 i를 넣어줌.

answer_list =[]
for i in range(M): # 문제 풀기
  problem = input()
  if problem.isdigit(): # 입력값이 숫자인 경우
    answer_list.append(pokemon_dict2[int(problem)])
  else: # 입력값이 문자인 경우
    answer_list.append(str(pokemon_dict1[problem]))

print('\n'.join(answer_list))
```
