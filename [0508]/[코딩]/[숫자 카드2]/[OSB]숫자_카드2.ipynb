* 위 코드는 시간초과 문제
* 문제의 순서가 출력에서 반영되므로 Set를 사용X
  * dictionary 사용

```python
N = int(input())
N_list = list(input().split(" "))
N_dict = {} # 딕셔너리 선언
for i in N_list:
  temp = 0
  for j in N_list:
    if i == j:
      temp += 1
  N_dict[i] = str(temp) # dictionary에 추가

M = int(input())
M_set = list(input().split(" "))
Answer_list = [N_dict[i] if i in N_dict else str(0) for i in M_set]
print(' '.join(Answer_list))
```

* 위 코드도 시간 초과
* `sys.stdin.readline()`사용
* `sys.stdout.write(str(data))`이용

```python
import sys 
N = int(sys.stdin.readline())
N_list = list(sys.stdin.readline().split(" "))
N_dict = {}
for i in N_list:
  temp = 0
  for j in N_list:
    if i == j:
      temp += 1
  N_dict[i] = str(temp) # dictionary에 추가

M = int(sys.stdin.readline())
M_list = list(sys.stdin.readline().split(" "))
Answer_list = [N_dict[i] if i in N_dict else str(0) for i in M_set]
for data in Answer_list:
    sys.stdout.write(data)
```

* 그래도 시간 초과
* for문 x for문이므로 실제 코드 상에서는 $N^2$ 가 되어 이를 단순히 for문 한개만 사용해서 구현.

```python
import sys
N = int(sys.stdin.readline())
N_list = list(map(int, sys.stdin.readline().split()))
N_dict = {}
for num in N_list:
    if num in N_dict: # 있으면 처음에는 1 그 뒤로는 += 1로 개수대로 세기 사용
        N_dict[num] += 1
    else:
        N_dict[num] = 1

M = int(sys.stdin.readline())
M_list = list(map(int, sys.stdin.readline().split()))
Answer_list = [str(N_dict[i]) if i in N_dict else '0' for i in M_list]
sys.stdout.write(' '.join(Answer_list))
```
