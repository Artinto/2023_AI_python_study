# 예제 
image.png

```python
N, M = map(int, input().split(" "))
total = N+M
S = set() # 집합 S에 포함되어 있는 문자열
check_list = [] # 검사할 문자열
for i in range(total):
  if i < N:
    S.add(input())
  else:
    check_list.append(input())

# 검사
Answer = ['1'  for i in check_list if i in S] # 맞으면 데이터 추가 아니면 pass
print(len(Answer))
```
