```python
n = int(input())
log = set() # 기록을 위한 집합 생성
for i in range(n):
  name, state = input().split(" ") # 이름과 state를 분리해서 받아드림.
  log.add(name)
  if state == 'leave': # 나가면 제거
    log.remove(name)
print("\n".join(sorted(log, reverse =True))) # 역순으로 한 줄에 한 명씩 출력
```

* 위 코드의 채점 시간이 너무 느려서 다른 방법을 생각
  * 원인 remove의 경우 set이 많아질 수록 더 오래 걸리게 됨.
  * 해결방법은 Dictionary의 del 기능 사용.

```python
n = int(input())
log = {} # 기록을 위한 딕셔너리 생성
for i in range(n):
  name, state = input().split(" ") # 이름과 state를 분리해서 받아드림.
  if state == 'enter': # 들어올 때마다 기록에 추가
    log[name] = True # 이름과 True를 넣어줌.
  else: # 나갈 때는 삭제
    del log[name]
print("\n".join(sorted(log.keys(), reverse=True))) # 역순으로 한 줄에 한 명씩 출력
```
