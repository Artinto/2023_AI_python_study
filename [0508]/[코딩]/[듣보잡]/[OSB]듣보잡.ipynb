```python
import sys
N, M = map(int, sys.stdin.readline().split())
N_set = set() # 듣도 못한 사람 
M_set = set() # 보도 못한 사람
for i in range(N):
  N_set.add(sys.stdin.readline().rstrip()) # 개행 문자 제거 후 집합에 넣기
for i in range(M): 
  M_set.add(sys.stdin.readline().rstrip()) # 개행 문자 제거 후 집합에 넣기

intersection = sorted(list(N_set & M_set)) # 듣도 보도 못한 사람(교집합)
print(len(intersection)) # 길이 출력을 통해 개수
print('\n'.join(intersection)) # 그리고 교집합 출력.
```

```python
N, M = map(int, input().split(" "))
N_set = set() # 듣도 못한 사람 
M_set = set() # 보도 못한 사람
for i in range(N):
  N_set.add(input()) # 각 집합에 넣어주기
for i in range(M): 
  M_set.add(input()) # 각 집합에 넣어주기

intersection = sorted(list(N_set & M_set)) # 듣도 보도 못한 사람(교집합)
print(len(intersection)) # 길이 출력을 통해 개수
print('\n'.join(intersection)) # 그리고 교집합 출력.
```
