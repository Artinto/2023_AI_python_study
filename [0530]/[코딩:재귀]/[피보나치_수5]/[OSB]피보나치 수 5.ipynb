* 재귀함수

```python
def fibonacci(n):
  # 처음 두개만 숫자로 넣어주고
  if n == 0:
    return 0
  elif n ==1:
    return 1
  return fibonacci(n-1) + fibonacci(n-2) # n이 2이상일떄부터 진행.

n = int(input())
print(fibonacci(n))
```
