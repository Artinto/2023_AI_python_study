```python
n = int(input())
```

```python
def hanoi(n, start, mid, end): # 원판 개수와 각 장대 입력 받기(순서 출력을 위해)
    if n == 1: #(n=1 일떄는 원판이 1개이므로 특수경우.)
        print(start, end)
    else:
        hanoi(n-1, start, end, mid) # 작은 건 중간으로 옮김. 순서를 꼬아서 n이 짝수 홀수인것도 반영되게 함. 
        print(start, end) # 옮긴 과정 출력.
        hanoi(n-1, mid, start, end) # 이제 즁엉애서 큰 원판을 끝으로 옮기는 과정.. 

n = int(input())
move_cnt = 2**n - 1
print(move_cnt)
hanoi(n, 1, 2, 3)
```
