```python
def star(x,y, n, matrix): # x, y 좌표와 크기, 출력용 리스트 받기
  # n =3일 떄의 ***
  #             * *
  #             *** 가 출력되도록함.
  if n == 1:
    matrix[x][y] = '*'
    return

  else:
    m = n//3
    for i in range(3):
      for j in range(3):
        if i == 1 and j == 1: # 가운데의 경우는 공백.
          pass
        else:
          star(x + i*m, y + j*m, m, matrix)  # 재귀 호출 각 자표에 i*m, j*m를 통해서 중간의 공백을 나타낼 수 있음.

n = int(input())
matrix = [[' ' for _ in range(n)] for _ in range(n)]  # n x n 크기의 공백의 리스트 생성
star(0, 0, n, matrix) # 호출.

# 결과를 한줄씩 출력해서 join함수 사용 가능하도록 함.
for row in matrix:
  print(''.join(row)) # join이용.
```
