* 재귀함수 이용.

```python
n =  int(input())

def factorial(x):
  if x == 0: # n이 0이면 1을 반환하고 끝냄.
    return 1
  temp = x * factorial(x-1)
  return temp
if n <= 20 and n >=0:
  print(factorial(n))
```
