* 알고리즘 수열 - 병합 정렬과 매우 유사한 문제.

```python
def str_split(str_data):
  l = len(str_data)+1
  if l <= 2:
    return str_data
  mid = l // 2 # 정 가운데 찾기
  first = str_split(str_data[:(l // 3)]) # 3등분
  second = str_split(str_data[(l // 3):2*(l // 3)]) # 3등분
  third = str_split(str_data[2*(l//3):3*(l // 3)]) # 3등분
  output = []
  f_index, s_index, t_index = 0,0,0 # index 정의
  while f_index<len(first):
    output.append(first[f_index])
    f_index += 1
  while s_index<len(second):
    output.append(' ')
    s_index += 1
  while t_index<len(third):
    output.append(third[t_index])
    t_index += 1 
  return output
```

```python
while True:
  try:
    n = int(input())
    str_data = list('-' * 3**n) 
    data  = str_split(str_data)
    print(''.join(data))
  except:
    break
```
