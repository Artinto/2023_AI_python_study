## 병합 정렬(Merge Sort)
* 배열을 2개의 조각으로 나눈후 각가을 정렬하고 그 조각들을 다시 모아 원래 배열을 정렬시키는 방법
* 보통 순환 호출을 이용해 부분 배열의 크기가 충분히 작아지도록 진행.
* 배열의 길이가 0 or 1이면 이미 정렬됬다고 판단.

__1. 단계__
  * 입력 배열을 비슷한 크기의 2개의 부분 배열로 분할.  

__2. 단계__
  * 부분 배열 정렬  

__3. 단계__
  * 정렬된 부분들을 하나의 배열로 합병


## 의사 코드

```python
def merge_sort(num_list, answer):#  0, 1
  if len(num_list) <= 1:
    return num_list
  q = (len(num_list)+1)//2
 # 차후 확인을 위해서 선언.
  # 분할.
  right =num_list[q:]
  left = num_list[:q]
  # 정렬
  left_sort = merge_sort(left, answer)
  right_sort = merge_sort(right, answer)
  return merge(left_sort, right_sort, answer)

def merge(left_sort, right_sort, answer):
  # 결과를 위한 최종 리스트 선언.
  output = [] 
  left_index = 0
  right_index = 0
  temp = 0
  while left_index < len(left_sort) and right_index < len(right_sort): # 들어온 두개의 배열이 배열의 끝을 가르키지 않을 경우..
    if left_sort[left_index] <= right_sort[right_index]: # 왼쪽의 요소가 더 작으면 추가
        output.append(left_sort[left_index])
        answer.append(left_sort[left_index])
        left_index += 1

    else:
        output.append(right_sort[right_index]) # 오른쪽의 요소가 더 작으면 추가
        answer.append(right_sort[right_index])
        right_index += 1
  # Merge 과정.(한쪽에 배열에 남아있는 원소가 있다면 결과에 추가)
  while left_index < len(left_sort):
    output.append(left_sort[left_index])
    answer.append(left_sort[left_index])
    left_index += 1
  while right_index< len(right_sort):
    output.append(right_sort[right_index])
    answer.append(right_sort[right_index])
    right_index += 1
  return output
```
