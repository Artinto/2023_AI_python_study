```python
def recursion(s, l, r, temp):
    temp += 1 # 얼마나 반환됬는지 알기 위해 선언.
    if l >= r: 
      return 1, temp
    elif s[l] != s[r]:  # 팰린드롬 판단 여부 확인.
      return 0, temp
    else: 
      return recursion(s, l+1, r-1, temp)

def isPalindrome(s, temp):
    return recursion(s, 0, len(s)-1, temp)
```

```python
num = int(input())
for i in range(num):
  test = input().rstrip()
  temp = 0
  answer = isPalindrome(test, temp)
  print(answer[0], answer[1])
```
