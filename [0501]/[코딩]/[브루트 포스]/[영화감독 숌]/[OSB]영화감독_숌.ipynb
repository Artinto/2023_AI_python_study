# 풀이
* 작은 수부터 나열을 해야하므로
  * 실제로 5666다음은 6666이 아니라 6660이된다. 
  * 6660 다음은 6661이 된다.

* 브루트포스
  * 무한반복문을 사용하여 구현.
  

```python
n = int(input())
cnt, num = 0, 666 # 시작할 숫자와 몇번째 요소인지를 알기 위해서 출력.
while True:
  if '666' in str(num): # 문자열로 반환해 666이 있는지 없는지를 판단
    cnt += 1
  if cnt == n:
    break
  num += 1
print(num)
```
