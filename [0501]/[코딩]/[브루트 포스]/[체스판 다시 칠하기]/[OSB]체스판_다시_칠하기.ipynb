# 풀이
* 처음에 행과 열이 주어지면, 해당 행에 맞게 BWBWBW 순서와 WBWBWB 순서의 행렬 생성
  * 그리고 그 이후 들어오는 행에 대해서 맞고 틀린것을 확인
  * 최소값 출력

```python
N, M = map(int, input().split())

# 판 기본판 생성 8 x 8
BW_board = []
for i in range(8):
  BW_string = ''
  for j in range(8):
    if (j+i)% 2 == 0:
      BW_string += 'B'
    else:
      BW_string += 'W'
  BW_board.append(BW_string)

WB_board = []
for i in range(8):
  WB_string = ''
  for j in range(8):
    if (j+i)% 2 == 0:
      WB_string += 'W'
    else:
      WB_string += 'B'
  WB_board.append(WB_string)

# 판 입력 받기
Board = [input() for i in range(N)]

# 비교 후 출력
WB_sum, BW_sum = [], []
for n in range(N-7):
  for m in range(M-7):
    WB, BW = 0, 0
    for i in range(8):
      for j in range(8):
        if BW_board[i][j] != Board[n+i][m+j]:
          BW +=1
        if WB_board[i][j] != Board[n+i][m+j]:
          WB +=1
    WB_sum.append(WB)
    BW_sum.append(BW)
if min(BW_sum) >= min(WB_sum):
  print(min(WB_sum))
else:
  print(min(BW_sum))
```
