# 풀이
* input으로 받아온 수를 list로 변환 및 sort 이용
  * map을 이용해 list와 int형으로 변환을 진행

```python
num = list(map(int, list(input())))
num.sort(reverse=True) # 내림차순으로 정렬
for i in num:
  print(i, end='') # 한줄로 출력을 위해서 end를 \n에서 ''로 변환
```
