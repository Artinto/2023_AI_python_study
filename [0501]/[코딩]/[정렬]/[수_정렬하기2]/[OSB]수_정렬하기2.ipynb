# 풀이
* 배열로 입력 받은 후 정렬
  * sort이용.

* 첫번쨰 코드(input함수가 너무 시간이 오래 걸림)
  * `sys.stdin.readline()` 이용
  * 출력 시 `sys.stdout.write(str(data)+ '\n')`이용

* `input() vs sys.stdin.readline()`
  * 입력 방식의 차이
    * input은 사용자에게 프롬프트를 출력해 사용자가 입력하도록함.
    * sys.stdin.readline()은 프롬프트를 출력하지 않고 받아서 Ente키를 이용해 종료점을 알려줌. 
  * 속도의 차이
    * input함수는 입력 값을 받을 때 마다 프롬프트를 출력하고, 입력 값을 입력할 때마다 문자열을 생성하므로, 반복문 안에서 `input`를 사용하게 되면 속도가 느려진다.
    * sys.stdin.readline() 한 번 호출할 때마다 한 줄의 문자열 반환하여 입력된 값을 한번의 처리가 가능해 빠른 속도로 동작할 수 있다.


```python
num = int(input())
num_list = [int(input()) for i in range(num)]
num_list.sort()
for data in num_list:
  print(data)
```

```python
import sys
#N개의 정수를 입력 받는다.
num = int(sys.stdin.readline())
num_list = [int(sys.stdin.readline()) for i in range(num)]
# 정렬
num_list.sort()
# 3. 정렬된 정수를 한 줄에 하나씩 출력한다.
for data in num_list:
    sys.stdout.write(str(data)+'\n')
```
