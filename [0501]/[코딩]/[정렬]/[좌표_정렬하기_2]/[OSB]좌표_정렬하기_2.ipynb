# 풀이 
 * 튜플 형식으로 list에 저장 후 sort
 * lambda 사용
  * lambda 매개변수 : 표현식구조
  * tuple의 두번째 요소, 첫번째 요소 순서로 
  * `lambda x:(x[1], x[0])`  
* input사용 대신 sys.stdin.readline() 이용   

[람다함수](https://wikidocs.net/22804)

```python
import sys
num = int(sys.stdin.readline())
a = [tuple(map(int, sys.stdin.readline().split())) for _ in range(num)]
a.sort(key=lambda x:(x[1], x[0])) # 두번째 요소, 첫번째 요소 순서로 정렬
for value in a:
    print(value[0], value[1])
```
