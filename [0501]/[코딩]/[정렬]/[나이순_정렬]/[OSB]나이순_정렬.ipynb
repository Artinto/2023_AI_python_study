# 풀이
 * 튜플 형식으로 list에 저장 후 sort
 * lambda 사용
  * lambda 매개변수 : 표현식구조
  * tuple의 첫번째 요소로 sort
    * 나이로만 sort가 되면 자동으로 나이가 같을 경우 먼저 가입한 순서대로 정렬됨.
  * `lambda x:(x[1])`  
* input사용 대신 sys.stdin.readline() 이용  

```python
import sys
num = int(sys.stdin.readline())
a = [tuple(sys.stdin.readline().split()) for _ in range(num)]
a.sort(key=lambda x:int(x[0])) # 첫 번째 요소인 나이를 기준으로 정렬 및 int형 으로 변환해서 sort
for value in a:
    print(value[0], value[1])
```
