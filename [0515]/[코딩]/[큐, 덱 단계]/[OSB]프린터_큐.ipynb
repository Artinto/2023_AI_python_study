```python
n = int(input())
for i in range(n):
  length, question = map(int, input().split())
  importance = list(map(int, input().split())) # 중요도 입력받기
  queue = [(importance[i], i) for i in range(length)] # tuple로(imporance와 index)로 생성
  count = 0
  while queue: # 중요도 함수가 다 사용되면, 그만
    temp = queue.pop(0) # 첫번째 값을 일단 pop
    if any(temp[0] < value[0] for value in queue): # 만약 위에서 추출한 것보다 조금이라도 높은 우선순위가 있다면 뒤로 보내기
      queue.append(temp)
    else: # 아니라면 (가장 높다는 의미)  
      count += 1 # 몇번 째로 출력됬는지 표시하기 위함.
      if question == temp[1]: # 뒤에 index가 질문과 같다면? 출력 후 종료
        print(count)
        break
```

* 다른 방법 찾아보기
  * deque모듈 이용
  * Deque(Double Ended Queue)
    * 양뱡향에서 데이터를 처리할 수 있는 자료 구조.
    * stack, queue 둘다 구조 만들 수 있음.
    * `deque.popleft`
      * 왼쪽 끝 element를 가져오고 삭제
    * `deque.pop`
      * 오른쪽 끝 element를 가져오고 삭제
    * `deque.append`
      * 오른쪽 끝에 element 추가
    * `deque.appendleft`
      * 왼쪽 끝에 element 추가
    * 리스트와 유사하지만, 양쪽에서 데이터를 넣어줄 수 있어서 범용성이 더 좋음.

```python
from collections import deque

n = int(input())
for i in range(n):
    length, question = map(int, input().split())
    importance = list(map(int, input().split())) # 중요도 입력받기
    queue = deque([(importance[i], i) for i in range(length)]) # deque에 튜플로(imporance와 index)로 생성
    count = 0
    while queue: # 중요도 함수가 다 사용되면, 그만
        temp = queue.popleft() # popleft는 fifo를 사용하겠다는 의미
        if any(temp[0] < value[0] for value in queue): # 만약 위에서 추출한 것보다 조금이라도 높은 우선순위가 있다면 뒤로 보내기
            queue.append(temp)
        else: # 아니라면 (가장 높다는 의미)  
            count += 1 # 몇번 째로 출력됬는지 표시하기 위함.
            if question == temp[1]: # 뒤에 index가 질문과 같다면? 출력 후 종료
                print(count)
                break
```
