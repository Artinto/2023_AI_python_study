```python
n, m = map(int, input().split(" "))
queue = [i for i in range(1, n+1)] # 1 ~ n까지
sequence_list = list(map(int, input().split(" "))) # 각 숫자를 법을 위치
count = 0
while sequence_list:
  temp = sequence_list.pop(0)
  while temp != queue[0]:
    a = count%len(queue) # 나머지 연산
    if temp in queue[1:1+len(queue)//2]: # 오른쪽에 있다면,
      queue.append(queue.pop(0)) # 왼쪽으로 이동.
    elif temp in queue[-len(queue)//2+1:]: # 왼편에 있다면,
      for _ in range(len(queue) -1): # 결과적으로 오른쪽으로 한칸 이동.
        queue.append(queue.pop(0)) # 왼쪽으로 이동
    count += 1
  queue.pop(0) # 빼버리기
print(count)
```

* deque 이용

```python
from collections import deque

n, m = map(int, input().split(" "))
queue = deque([i for i in range(1, n+1)]) # 1 ~ n까지
sequence_list = deque(map(int, input().split(" "))) # 각 숫자를 법을 위치
count = 0

while sequence_list:
    if sequence_list[0] == queue[0]:
        sequence_list.popleft()
        queue.popleft()

    else:
        target_idx = queue.index(sequence_list[0]) # 목표 인덱스
        left_move = target_idx # 왼쪽으로 이동하는 숫자카드 수
        right_move = len(queue)-target_idx # 오른쪽으로 이동하는 숫자카드 수

        if left_move <= right_move:
            queue.rotate(-left_move)
            count += left_move
        else:
            queue.rotate(right_move)
            count += right_move

print(count)
```
