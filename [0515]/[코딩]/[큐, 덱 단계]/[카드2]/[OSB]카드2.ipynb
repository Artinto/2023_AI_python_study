# 풀이
* 큐 구조
  * FIFO(First In Fist Out)의 구조로 먼저 집어 넣은 데이터가 먼저 나오는 구조이다.

```python
import sys
N = int(sys.stdin.readline().rstrip()) # 숫자 하나 입력 받기
int_queue = [i for i in range(1, N+1)]
while True:
  if len(int_queue) == 1: # 숫자가 하나만 남았을 경우 
    print(int_queue[0])
    break
  int_queue.pop(0) # 처음 pop
  int_queue.append(int_queue.pop(0)) # 그 다음 처음 pop 밑 맨 밑으로 옮기기
```

* 시간 초과 문제 발생
* pop을 사용하지 않는 방식으로 이용

```python
N = int(input()) 
int_queue = [i for i in range(1, N+1)]
front = 0 # 맨앞 
end = N-1 # 맨뒤
while end>front:
  #print(int_queue[front])
  front += 1 # 맨처음 pop 효과
  int_queue.append(int_queue[front]) # 맨 뒤로 보내기
  front += 1 # 맨 뒤로 보냇으므로 그 다음으로 이동해야함.
  end += 1 # 맨 뒤가 하나 늘어나는 효과
print(int_queue[front])
```

* 0, 5 -> 2, 6 -> 4, 7 -> 6, 8 -> 8, 9, -> 10, 10
* 1 -> 3 -> 5 -> 2 -> 6-> 4
