* Stack 구조 활용

```python
import sys
while True:
  str_input = input().rstrip() # \n 제거
  stack_list = []
  if str_input == '.': # 종료조건
    break
  elif str_input.strip() == '.':
    print('yes')
    continue
  else:
    check = 0
    for v in str_input:
      if v == '[' or v == '(':
        stack_list.append(v)
      elif v == ']':
        if len(stack_list) == 0 or stack_list[-1] != '[':
          check += 1
          break
        stack_list.pop() # 현재 스택의 마지막이 '[' 인경우
      elif v == ')':
        if len(stack_list) == 0 or stack_list[-1] != '(': 
          check += 1
          break
        stack_list.pop() # 현재 스택의 마지막이 '(' 인경우
          
    if check == 0 and len(stack_list) == 0:
      print('yes')
    else:
      print('no')
```
