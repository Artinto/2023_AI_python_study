```python
n = int(input())
# 순서가 필요하므로 list로 선언
stack_list = []

for _ in range(n):
  command = input().split() # 명렬을 받음.
  if command[0] == 'push':
    stack_list.append(int(command[1])) # push의 경우 commmand[1]이 존재.
  elif command[0] == 'pop':
    if not stack_list:
      print(-1)
    else:
      print(stack_list.pop())
  elif command[0] == 'size':
    print(len(stack_list))
  elif command[0] == 'empty':
    if not stack_list:
      print(1)
    else:
      print(0)
  elif command[0] == 'top':
    if not stack_list:
      print(-1)
    else:
      print(stack_list[-1])
```

```python
import sys
n = int(sys.stdin.readline().rstrip())
# 순서가 필요하므로 list로 선언 및 뒤에 pop를 사용하기 위해
stack_list = []
for _ in range(n):
  command = sys.stdin.readline().rstrip().split() #  push인 경우만 두개로 입력 받음.
  if command[0] == 'push':
    stack_list.append(int(command[1]))
  elif command[0] == 'pop':
    if not stack_list:
      print(-1)
    else:
      print(stack_list.pop())
  elif command[0] == 'size':
    print(len(stack_list))
  elif command[0] == 'empty':
    if not stack_list:
      print(1)
    else:
      print(0)
  elif command[0] == 'top':
    if not stack_list:
      print(-1)
    else:
      print(stack_list[-1])
```
