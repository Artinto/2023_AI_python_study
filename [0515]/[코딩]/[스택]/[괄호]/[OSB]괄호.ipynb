* Stack 처럼 사용.

```python
T = int(input())
for _ in range(T):
  vps_data = input()
  vps_list = []
  check = 0 # 중간 과정이 거쳐지면 1 아니면 0 중복 출력을 막기 위해서 사용.
  if len(vps_data) % 2 == 1: # 홀수 일경우 NO
    print('NO')
    continue
  else: # 짝수인 경우 중 
    for index, string in enumerate(vps_data):
      if string == '(':
        vps_list.append(string)
      elif string == ')':
        if len(vps_list) == 0:
          check += 1
          print('NO')
          break
        else:
          vps_list.pop() # 마지막 문제 제거
          if len(vps_list) == 0 and index == len(vps_data)-1:  # vps_llst가 0이면
            print('YES')
            check+=1
  if len(vps_data) > 0 and check == 0 :  # data가 있으면
    print('NO')
```
