* 마찬가지로 시간초과 문제 발생.
* 알고리즘을 다시 짜서 시간 복잡도를 낮춤.

```python
import sys
def sequence(input_list):
    output_str = ""   # 출력을 위한 str 자료
    stack = []        # 값을 저장할 스택
    
    num = 1  # 현재 숫자 (1 ~ N까지 증가)
    for i in input_list:
        # 스택이 비어있을 때
        while num <= i:
            stack.append(num)
            num += 1
            output_str += "+"
        # 스택의 맨 위의 값이 현재 입력된 값과 같을 때
        if stack[-1] == i:
            stack.pop()
            output_str += "-"
        # 불가능한 상황이면 NO 반환
        else:
            return "NO"
    return output_str
N = int(sys.stdin.readline().rstrip()) # 입력받기
input_list = [int(sys.stdin.readline().rstrip()) for _ in range(N)]
result = sequence(input_list) # 결과
if result == "NO":
    print(result)
else:
    for i in result:
        print(i)
```
