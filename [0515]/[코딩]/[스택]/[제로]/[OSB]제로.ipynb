```python
N = int(input())
answer_list = []
for _ in range(N):
  command = int(input())
  if command == 0:
    if len(answer_list) == 0:
      break # 이상한 경우 탈출
    else:
      answer_list.pop()
  else:
    answer_list.append(command)
print(sum(answer_list))
```

```python
import sys
n = int(sys.stdin.readline().rstrip())
answer_list = []
for _ in range(n):
  command = int(sys.stdin.readline().rstrip())
  if command == 0:
    if len(answer_list) == 0:
      break # 이상한 경우 탈출
    else:
      answer_list.pop()
  else:
    answer_list.append(command)
print(sum(answer_list))
```
